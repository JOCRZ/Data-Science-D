Linear regression is used to estimate the dependent variable in case of a change in independent variables. For example, predict the price of houses. Whereas logistic regression is used to calculate the probability of an event.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```

```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
def prediction_function(age):
    z = 0.11 * age - -4.54
    y = sigmoid(z)
    return y
```

```python
import pickle
model=pickle.load(open('model.pkl','rb'))
```

```python
def predict_forest(age):
    input=np.array([[age]]).astype(np.float64)
    prediction=model.predict(input)
    pred=prediction[0]
    if pred == '0':
            val = 'Will Not Buy Insurance'
            return val
    if pred == '1':
            val = 'Will Buy Insurance'
            return val
```

```python
def predict_buy(age):
    input=np.array([[age]]).astype(np.float64)
    prediction=model.predict(input)
    pred=prediction[0]
    return pred
```
